## Web-crawling Demographics Data

```python
from crawler import *
```

```python
index_html = fetch(DEMOGRAPHICS_INDEX)
```

```python
country_links = extract_country_links(index_html)
country_links
```

```python
records = {}
for country, url in tqdm(country_links.items(), desc="Crawling", unit="country"):
    html = fetch(url)
    records[country] = parse_country_page(html)
```

```python
df_demographics = pd.DataFrame.from_dict(records, orient="index")
df_demographics.index.name = "Country"
df_demographics
```
